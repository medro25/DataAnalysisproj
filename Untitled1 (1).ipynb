```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName('YourAppName') \
    .config('spark.network.timeout', '1000s') \
    .getOrCreate()

```

```python
df = spark.read.csv("Production_Crops_Livestock_E_All_Data.csv", header=True, inferSchema=True)
#df.show(vertical=True)
```

```python
from pyspark.sql.functions import col

columns = ['Item', 'Element', 'Area', 'Y2000', 'Y2005', 'Y2010', 'Y2015', 'Y2020']
df_filtered = df.select(*columns).filter((col('Item').isin(['Maize (corn)', 'Rice', 'Wheat', 'Barley', 'Sorghum'])) & (col('Element') == 'Production'))

df_filtered.show()


```

```python


from pyspark.sql.functions import sum

df_sum = df_filtered.groupby('Item').agg(
    sum('Y2000').alias('Y2000'),
    sum('Y2005').alias('Y2005'),
    sum('Y2010').alias('Y2010'),
    sum('Y2015').alias('Y2015'),
    sum('Y2020').alias('Y2020')
)

df_sum.show()


```

```python
from pyspark.sql.functions import col, expr

# Select the relevant columns and filter the dataframe
columns = ['Item', 'Element', 'Area', 'Y2000', 'Y2005', 'Y2010', 'Y2015', 'Y2020']
df_filtered = df.select(*columns).filter((col('Item').isin(['Maize (corn)', 'Rice', 'Wheat', 'Barley', 'Sorghum'])) & (col('Element') == 'Production'))

# Pivot the dataframe to transpose the rows and columns
df_transposed = df_filtered.selectExpr(
    "Item",
    "stack(5, 'Y2000', Y2000, 'Y2005', Y2005, 'Y2010', Y2010, 'Y2015', Y2015, 'Y2020', Y2020) as (Year, Production)"
).groupBy("Year").pivot("Item").sum("Production").orderBy("Year")

df_transposed.show()

```

```python
import matplotlib.pyplot as plt
import pyspark.sql.functions as F
from matplotlib.ticker import FuncFormatter, ScalarFormatter

# Pivot the dataframe to transpose the rows and columns
df_transposed = df_filtered.selectExpr(
    "Item",
    "stack(5, 'Y2000', Y2000, 'Y2005', Y2005, 'Y2010', Y2010, 'Y2015', Y2015, 'Y2020', Y2020) as (Year, Production)"
).groupBy("Year").pivot("Item").sum("Production").orderBy("Year")
print(df_transposed)

# Extract the columns from the pivot table
years = [row.Year for row in df_transposed.select('Year').collect()]
maize = [row['Maize (corn)'] for row in df_transposed.select('Maize (corn)').collect()]
rice = [row['Rice'] for row in df_transposed.select('Rice').collect()]
wheat = [row['Wheat'] for row in df_transposed.select('Wheat').collect()]
barley = [row['Barley'] for row in df_transposed.select('Barley').collect()]
sorghum = [row['Sorghum'] for row in df_transposed.select('Sorghum').collect()]

# Plot the line chart
plt.plot(years, maize, label='Maize (corn)')
plt.plot(years, rice, label='Rice')
plt.plot(years, wheat, label='Wheat')
plt.plot(years, barley, label='Barley')
plt.plot(years, sorghum, label='Sorghum')

# Add title and axis labels
plt.title("Production of Crops by Year")
plt.xlabel("Year")
plt.ylabel("Production (tonnes)")

# Set the y-axis formatter to format the values as integers
plt.gca().get_yaxis().set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))

# Set the y-axis tick label format to scientific notation
fmt = ScalarFormatter(useMathText=True)
fmt.set_scientific(True)
fmt.set_powerlimits((-3,3))
plt.gca().yaxis.set_major_formatter(fmt)

# Add legend and show the plot
plt.legend()
plt.show() 
```

```python
df = spark.read.csv("Production_Crops_Livestock_E_All_Data.csv", header=True, inferSchema=True)
#df.show(vertical=True)
dg = df.limit(61273)

# Display the filtered dataframe
dg.show()
```

```python
from pyspark.sql.functions import col, expr
from pyspark.sql.window import Window
from pyspark.sql.functions import coalesce
from pyspark.sql import functions as F

# Select the relevant columns and filter the dataframe
columns = ['Item', 'Element', 'Area', 'Y2000', 'Y2005', 'Y2010', 'Y2015', 'Y2020']
dg_filtered = dg.select(*columns).filter((col('Item').isin(['Maize (corn)', 'Rice', 'Wheat', 'Barley', 'Sorghum'])) & (col('Element') == 'Production'))
#dg_filtered.show()
dg_filtered_sumlargest = dg_filtered.groupBy('Item', 'Area').agg(F.sum('Y2020').alias('Y2020')).orderBy(F.desc('Y2020'))

# Show the resulting DataFrame
#dg_filtered_sumlargest.show()
# Create a window specification to partition by 'Item' and order by descending 'Y2020' sum
windowSpec = Window.partitionBy('Item').orderBy(col('Y2020').desc())

# Assign rank to each row within the partition
dg_filtered_sumlargest = dg_filtered_sumlargest.withColumn('rank', expr('rank() over (partition by Item order by Y2020 desc)'))

# Filter for the top 3 countries for each item based on the sum values
top_countries = dg_filtered_sumlargest.filter(col('rank') <= 3)

# Show the resulting DataFrame
top_countries.show()
# Sum up the values for countries not in the top 3 and name them as "Other" for each crop
other_countries = dg_filtered_sumlargest.filter(col('rank') > 3).groupBy('Item').agg(expr('sum(Y2020) as Y2020')).withColumn('Area', expr("'Other'"))

# Union the top countries with the "Other" countries
result = top_countries.select('Item', 'Area', 'Y2020').union(other_countries.select('Item', 'Area', 'Y2020'))
result.show()
# Filter for the rows with 'Area' as "Other"
result_filtered = result.filter(col('Area') == 'Other')

# Show the resulting DataFrame
result_filtered.show()
#transform the data 
dg_plot= result.toPandas()
# Pivot the dataframe to have crops as rows and countries as columns
dg_plot = dg_plot.pivot(index='Item', columns='Area', values='Y2020').fillna(0)

print(dg_plot)
# Convert the values to percentages
dg_plot = dg_plot.apply(lambda x: x/x.sum()*100, axis=1)
print("lambda",dg_plot)
# Sort the rows by total value (sum of top 4 countries and "Other")
dg_plot = dg_plot.loc[dg_plot.sum(axis=1).sort_values(ascending=False).index]

# Plot the stacked bar chart
ax = dg_plot.plot(kind='bar', stacked=True, figsize=(10, 6),legend=None)
plt.xlabel('Crop')
plt.ylabel('Percentage')
plt.title('Top 3 countries and Other for each crop in 2020')
#plt.legend(title='Country', loc='upper left')

# Add country labels inside the bars
for i, (crop, row) in enumerate(dg_plot.iterrows()):
    offset = 0
    for country, value in row.iteritems():
        if value != 0:
            ax.text(i, offset + value/2, country, ha='center', va='center')
            offset += value

plt.show()


```
